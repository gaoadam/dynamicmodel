# Import Libraries

```python
#Import source code
from src import dynamicmodel as dm

#Numeric/Computational libraries
import numpy as np
from scipy import signal
import time

#Visualization libraries
from matplotlib import pyplot as plt
from matplotlib import cm
from matplotlib import colors

#Formatting libraries
import pandas as pd

#Path management library
from pathlib import Path
```

# Generate Signals

```python
#Initialize parameters
dt =.02
N = 5000

#Initialize u at t=0
x_t0 = np.array([0,0.05])
```

```python
x_array_dampeddriven = dm.x_driven(x_t0=x_t0, dt=dt, N=N, m=1, b=0.1, k=1, u1=dm.harmonics, args={'n_list':[1], 'a_list':[1]})
dm.plot_x(x_array_dampeddriven, columns=['t','x','velocity'], title='Damped Oscillator')
```

```python
x_array_rlc = dm.x_rlc(x_t0=x_t0, dt=dt, N=N, r=1, l=1, c=1, u1=dm.harmonics, args={'n_list':[1], 'a_list':[1]})
dm.plot_x(x_array_rlc, columns=['t','capacitor voltage', 'inductor current'], title='RLC Circuit')
```

```python
#Configure save paths
save_folder = "exports"
save_path = "{}\\{}\\".format(str(Path.cwd()), save_folder)

#Save files
pd.DataFrame(data=x_array_dampeddriven, columns=['t','x', 'v']).to_csv(save_path+'x_dampeddriven.csv', index=False)
pd.DataFrame(data=x_array_rlc, columns=['t','capacitor voltage', 'inductor current']).to_csv(save_path+'x_rlc.csv', index=False)
```
